**Generation block without parameters**

```python
def generate():
    messages = [
       {"role":"system","content":"you are an helpful assistant you will help me to write an sorry essay that i did accident"
       },
       {"role":'user',"content":"write an sorry essay that i did accident by mistake"}
    ]
    response = client.chat.completions.create(
        model=FREE_MODEL,
        messages=messages
    )
    return response.choices[0].message.content
result = generate()
print(result,end=" ")
```

**REFLECTION_PATTERN**

```python
def reflect(generated_text,system_prompt):
  messages=[
      {"role":"system","content":system_prompt},
      {"role":"user","content":generated_text}

  ]
  response = client.chat.completions.create(
      model=FREE_MODEL,
      messages=messages
  )
  return response.choices[0].message.content

def reflection_loop(max_turns=3):

    generated = generate(
        "You are an expert dietitian.",
        "Give me a diet plan to lose 10kg in 2 months. I am 5'11 and 98kg."
    )

    for i in range(max_turns):
        print(f"\n--- Turn {i+1} ---")
        print("Generated:\n", generated)

        feedback = reflect(generated, "You are a strict critic. Give feedback. If perfect say LGTM.")
        print("\nFeedback:\n", feedback)

        # Modified condition to check for exact 'LGTM' as the acceptance signal
        if feedback.strip() == "LGTM":
            print("✅ Done!")
            break

        generated = generate(
            "You are an expert dietitian.",
            f"Improve this based on feedback.\n\nPlan:\n{generated}\n\nFeedback:\n{feedback}"
        )

    return generated

reflection_loop()
```
